```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("data/virus_titers.csv")
```

```python

virus_order = ["WT", "D77E", "R149T", "S136H", "I155H"]

base = (
    alt.Chart(df)
    .mark_point(size=300, filled=True, stroke='white')
    .encode(
        x=alt.X(
            "cells titrated on:N",
            title=None,
            axis=alt.Axis(
                labelAngle=45,
                labelFontSize=16,
                labelLimit=1000,   # prevent truncation
                grid=False,
            ),
        ),
        y=alt.Y(
            "TU/ml:Q",
            title="TU/ml",
            scale=alt.Scale(type="log"),
            axis=alt.Axis(
                labelFontSize=16,
                titleFontSize=22,
                grid=False,
                format=".0e",
            ),
        ),
        color=alt.Color(
            "cells produced in:N",
            title="Cells produced in",
            legend=alt.Legend(
                orient="bottom",
                direction="horizontal",
                columns=1,          # stack entries so full text fits
                labelLimit=1000,    # prevent legend truncation
                titleLimit=1000,
                titleFontSize=20,
                labelFontSize=16,
                symbolSize=180,
            ),
        ),
        tooltip=[
            alt.Tooltip("virus:N"),
            alt.Tooltip("cells titrated on:N"),
            alt.Tooltip("cells produced in:N"),
            alt.Tooltip("TU/ml:Q", format=".2e"),
        ],
    )
    .properties(width=220, height=260)
)

facet_chart = base.facet(
    column=alt.Column(
        "virus:N",
        sort=virus_order,
        title=None,
        header=alt.Header(
            labelFontSize=18,
            titleFontSize=18,
        ),
    )
).resolve_scale(y="shared")

x_title = (
    alt.Chart(pd.DataFrame({"text": ["Cells titrated on"]}))
    .mark_text(fontSize=24, fontWeight="bold", align="left")
    .encode(text="text:N")
    .properties(height=40, width=1200)
)

chart = (
    alt.vconcat(
        facet_chart,
        x_title,
        spacing=10,
    )
    .configure_view(stroke=None)
    .configure_axis(
        labelFontSize=16,
        titleFontSize=20,
    )
    .configure_legend(
        labelLimit=1000,
        titleLimit=1000,
    )
    .configure(
        padding={"left": 10, "right": 40, "top": 10, "bottom": 20}
    )
)

chart
```

```python
# ── Save chart as html ─────────────────────────────────────────────────────────
OUTPUT_html= './conditional_flu_virus_titers.html'
# save as HTML
chart.save(OUTPUT_html)
print(f"✓ Chart saved as HTML: {OUTPUT_html}")
```
